# Data Visualization
It's an art of representing raw data in the form of beautiful graphs.

These are tools for data visualization
- pandas 
- matplotlib(common)
- seaborn(common)
- ggplot
- plotly

working with so many raw data. so finding insights of these data is extremely hard . so data visualization help us

Why?
- Easy to understand graphs
- better analysis
- detecting errors
- efficient
- good for storytelling
- quick decision making
- identify patterns
- exploring insights

Types:-

    - Line Graph(-mostly used for short periods
                - used to track changes over short and long periods of time.
                - also used for more than one group.
                - we can compare trends between two or more than two groups.
                - also show the relationship between groups.)
    
    - Bar Graph(- mostly used for long periods of time.)
    
    - boxplot(-mostly used for detecting the outliers.
             - it also divides the data into sections like 25%,50%,75% , it also show us the max and min value.
             - it also help us to find mean of data quickly.)
    
    - Histogram(-it also groups the data into bins.
               - used for visualizing both discrete data or continous data.
               - it also measures the interval of time using bins.
               - also used this for frequency distribution.)
    
    - Pie Chart(- show percentage of whole data and represent % at a set point in time.
               - it don't changes w.r.t time.)
                
    - Scatter plot(-used for relation and co-relation between each other or not.)
    -etc.

# matplotlib

```python
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
plt.title("sepal length")
plt.xlabel("index")
plt.ylabel("sepal length values")
plt.plot(df['sepal.length'])
plt.legend("sepal length")
plt.show()
```

```python
plt.title("sepal length")
plt.xlabel("index")
plt.ylabel("sepal length values")
plt.plot(df['sepal.length'],'--r')
plt.legend("sepal length")
plt.show()
```

```python
plt.title("comparing sepal length and petal length")
plt.xlabel("index")
plt.ylabel("sepal length values")
plt.plot(df['sepal.length'],'--r')
plt.plot(df['petal.length'],'--g')
plt.legend(['sepal length','petal length'])
plt.show()
```

```python
plt.title("comparing sepal length and petal length")
plt.xlabel("index")
plt.ylabel("sepal length values")
plt.plot(df['sepal.length'],'--r')
plt.plot(df['petal.length'],'--g')
plt.xticks(rotation=60)
plt.yticks(rotation=45)
plt.legend(['sepal length','petal length'])
plt.show()
```
